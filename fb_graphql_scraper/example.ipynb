```python
#%% pip install facebook-graphql-scraper
from fb_graphql_scraper.facebook_graphql_scraper import FacebookGraphqlScraper as fb_graphql_scraper
from IPython.display import display
import pandas as pd
import pprint
```

# Example.1 - without logging in

- setting parameter

```python
facebook_user_name = "KaiCenatOfficial" # target account
days_limit = 60 # Number of days within which to scrape posts
driver_path = "/Users/renren/Downloads/chromedriver-mac-arm64 2/chromedriver" # driver path
```

- collect posts

```python
fb_spider = fb_graphql_scraper(driver_path=driver_path)
res = fb_spider.get_user_posts(
    fb_username_or_userid=facebook_user_name, 
    days_limit=days_limit,
    display_progress=True
)
```

- show result

```python
pprint.pp(res)
```

```python
df = pd.json_normalize(res['data'])
display(df)
```
